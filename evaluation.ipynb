```python
import orchest
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score,accuracy_score,classification_report,confusion_matrix
```

```python
data = orchest.get_inputs()
lgbm, pred_lgbm = data["lgbm"]
rf, pred_rf, X_train, y_train, X_val, y_val, test = data["RF"]
nb, pred_nb = data["NB"]
lr, pred_lr = data["LR"]
total_score=[]
```

# LGBM

```python
lgbm_f1 = f1_score(y_val, pred_lgbm)
lgbm_score = accuracy_score(y_val, pred_lgbm)
print("LGBM F1 score:", lgbm_f1)
print("LGBM Accuracy Score:",lgbm_score)
total_score.append(['LGBM',lgbm_f1,lgbm_score])
```

```python
plt.figure(figsize=(10, 10))
cm = confusion_matrix(y_val, pred_lgbm)
sns.heatmap(
    cm,
    cmap="Blues",
    linecolor="black",
    linewidth=1,
    annot=True,
    fmt="",
    xticklabels=["Disaster", "Non_Disaster"],
    yticklabels=["Disaster", "Non_Disaster"],
)
plt.xlabel("Predicted")
plt.ylabel("Actual");
```

# RF

```python
RF_f1 = f1_score(y_val, pred_rf)
RF_score = accuracy_score(y_val, pred_rf)
print("RF F1 score:", RF_f1)
print("RF Accuracy Score:",RF_score)
total_score.append(['RF',RF_f1,RF_score])
```

```python
plt.figure(figsize=(10, 10))
cm = confusion_matrix(y_val, pred_rf)
sns.heatmap(
    cm,
    cmap="Blues",
    linecolor="black",
    linewidth=1,
    annot=True,
    fmt="",
    xticklabels=["Disaster", "Non_Disaster"],
    yticklabels=["Disaster", "Non_Disaster"],
)
plt.xlabel("Predicted")
plt.ylabel("Actual");
```

# LR

```python
LR_f1 = f1_score(y_val, pred_lr)
LR_score = accuracy_score(y_val, pred_lr)
print("LR F1 score:", LR_f1)
print("LR Accuracy Score:",LR_score)
total_score.append(['LR',LR_f1,LR_score])
```

```python
plt.figure(figsize=(10, 10))
cm = confusion_matrix(y_val, pred_lr)
sns.heatmap(
    cm,
    cmap="Blues",
    linecolor="black",
    linewidth=1,
    annot=True,
    fmt="",
    xticklabels=["Disaster", "Non_Disaster"],
    yticklabels=["Disaster", "Non_Disaster"],
)
plt.xlabel("Predicted")
plt.ylabel("Actual");
```

# NB

```python
NB_f1 = f1_score(y_val, pred_nb)
NB_score = accuracy_score(y_val, pred_nb)
print("NB F1 score:", NB_f1)
print("NB Accuracy Score:",NB_score)
total_score.append(['NB',NB_f1,NB_score])
```
